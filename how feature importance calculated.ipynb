```python
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier
```

```python
X,y=make_classification(n_samples=10,n_features=2,n_informative=2,n_redundant=0,n_classes=2,random_state=2)
```

```python
X
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(X,y)
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(clf)
```

```python
clf.feature_importances_
```

```python
X,y=make_classification(n_samples=10,n_features=3,n_informative=2,n_redundant=0,n_classes=2,random_state=2)
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(X,y)
```

```python
plot_tree(clf)
```

```python
clf.feature_importances_
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=2)
```

```python
rf.fit(X,y)
```

```python
print(rf.estimators_[0].feature_importances_)
print(rf.estimators_[1].feature_importances_)
```
