```python
upper_limit=df['Height'].quantile(0.99)
lower_limit=df['Height'].quantile(0.01)
```

```python
print("upper limit:",upper_limit)
print("lower limit",lower_limit)
```

```python
new_df=df[(df['Height']>=lower_limit) & (df['Height']<=upper_limit)]
```

```python
df[(df['Height']<=lower_limit) | (df['Height']>=upper_limit)]
```

```python
new_df['Height'].describe()
```

```python
sns.distplot(new_df['Height'])
```

```python
sns.boxplot(x=new_df['Height'])
```

# Capping

```python
#capping-->winserization
```

```python
new_df_cap=df.copy()
```

```python
new_df_cap['Height']=np.where(new_df_cap['Height']>=upper_limit,upper_limit,np.where(new_df_cap['Height']<=lower_limit,lower_limit,new_df_cap['Height']))
```

```python
new_df_cap.describe()
```

```python
sns.distplot(new_df_cap['Height'])
```

```python
sns.boxplot(x=new_df_cap['Height'])
```
