```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import plotly.offline as pyo
```

```python
match=pd.read_csv('matches.csv')
delivery=pd.read_csv('deliveries.csv')
```

```python
match.head()
```

```python
delivery.head()
```

```python
ipl=delivery.merge(match,left_on='match_id',right_on='id')
```

```python
ipl.head()
```

```python
ipl.shape
```

```python
#scatter plots are drawn between two continuous variables
#we are going to draw scatter plots between batsman avg(x_axis) and batsman strike rate(y_axis) of top 50 batsman of all time
```

```python
#avg vs sr of top 50 batsman
```

```python
#fetching a new dataframe of top 50 batsman
```

```python
top50=ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(50).index.tolist()
```

```python
new_ipl=ipl[ipl['batsman'].isin(top50)]
```

```python
new_ipl
```

```python
#strike rate=no of runs scored/no of balls played
runs=new_ipl.groupby('batsman')['batsman_runs'].sum() #total run scored by a player
balls=new_ipl.groupby('batsman')['batsman_runs'].count() #total no of ball played bt batsman
sr=(runs/balls)*100
sr=sr.reset_index()
sr
```

```python
#calulating avg
#avg=total no of runs/total no of outs
out=ipl[ipl['player_dismissed'].isin(top50)]
```

```python
n_outs=out['player_dismissed'].value_counts()
```

```python
avg=runs/n_outs
```

```python
avg=avg.reset_index()
```

```python
avg
```

```python
avg.rename(columns={'index':'batsman',0:'avg'},inplace=True)
```

```python
avg
```

```python
avg=avg.merge(sr,on='batsman')
```

```python
avg
```

```python
#plot scatter plot here
trace=go.Scatter(x=avg['avg'],y=avg['batsman_runs'],mode='markers',text=avg['batsman'],marker={'color':'#00a65a','size':16})

data=[trace] #there will be multiple traces as our requirements
layout=go.Layout(title='Batsman Avarage vs Batsman strike rate',
                xaxis={'title':'Batsman Avarage'},
                yaxis={'title':'Batsman Strike Rate'})


fig=go.Figure(data=data,layout=layout)
pyo.plot(fig,filename='Scatter.html')
```

```python
#Multiple line charts
def batsman_comp(*name):
    data=[]
    for i in name:
        target=ipl[ipl['batsman']==i]
        performance=target.groupby('season')['batsman_runs'].sum().reset_index()
        trace=go.Scatter(x=performance['season'],y=performance['batsman_runs'],mode='lines+markers',marker={'size':16},name=i)
        data.append(trace)
    layout=go.Layout(title="Year by year performance of batsmans",
                xaxis={'title':'Season'},
                yaxis={'title':'Total runs'})
    fig=go.Figure(data=data,layout=layout)
    pyo.plot(fig,filename='Year by Year')
```

```python
batsman_comp('V Kohli','MS Dhoni','AB de Villiers','SK Raina','G Gambhir')
```

# Box plot

```python
match_agg=delivery.groupby('match_id')['total_runs'].sum().reset_index()
```

```python
season_wise=match_agg.merge(match,left_on='match_id',right_on='id')[['match_id','season','total_runs']]
```

```python
season_wise
```

```python
trace=go.Box(x=season_wise['total_runs'],name='All_seasons',marker={'color':'#00a65a'})
data=[trace]
layout=go.Layout(title='Total score Analysis',
                xaxis={'title':'Total Score'})
fig=go.Figure(data=data,layout=layout)
pyo.plot(fig)
```

```python
mask1=season_wise['season']==2017
mask2=season_wise['season']==2008
```

```python
season_2017=season_wise[mask1]
season_2008=season_wise[mask2]
```

```python
trace1=go.Box(x=season_2017['total_runs'],name='2017',marker={'color':'#00a65a'})
trace2=go.Box(x=season_2008['total_runs'],name='2008')
data=[trace1,trace2]
layout=go.Layout(title='Total score Analysis',
                xaxis={'title':'Total Score'})
fig=go.Figure(data=data,layout=layout)
pyo.plot(fig)
```
