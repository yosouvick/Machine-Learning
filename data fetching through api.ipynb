# Empty dataframe


```python
df=pd.DataFrame()
```

```python
df
```

```python
for i in range(1,551):
    response=requests.get(f"https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={i}")
    temp_df=pd.DataFrame(response.json()['results'])[['id','original_title','overview','popularity','release_date','vote_average','vote_count']]
    df=df.append(temp_df,ignore_index=True)
```
