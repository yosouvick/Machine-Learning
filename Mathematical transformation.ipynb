```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeClassifier
import seaborn as sns
```

```python
df=pd.read_csv('train.csv',usecols=['Age','Fare','Survived'])
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:,1:3]
y=df.iloc[:,0]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
sns.distplot(x_train['Age'])
plt.title("Age PDF")
plt.subplot(122)
stats.probplot(x_train['Age'],dist="norm",plot=plt)
plt.title("Age QQplot")
plt.show()
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
sns.distplot(x_train['Fare'])
plt.title("Fare PDF")
plt.subplot(122)
stats.probplot(x_train['Fare'],dist="norm",plot=plt)
plt.title("Fare QQplot")
plt.show()
```

```python
clf1=LogisticRegression()
clf2=DecisionTreeClassifier()
```

```python
clf1.fit(x_train,y_train)
clf2.fit(x_train,y_train)
y_pred1=clf1.predict(x_test)
y_pred2=clf2.predict(x_test)
print("Accuracy LR",accuracy_score(y_test,y_pred1))
print("Accuracy DT",accuracy_score(y_test,y_pred2))
```

```python
trf=FunctionTransformer(func=np.log1p)
```

```python
x_train_tranformed=trf.fit_transform(x_train)
x_test_tranformed=trf.transform(x_test)
```

```python
clf1=LogisticRegression()
clf2=DecisionTreeClassifier()
clf1.fit(x_train_tranformed,y_train)
clf2.fit(x_train_tranformed,y_train)
y_pred1=clf1.predict(x_test_tranformed)
y_pred2=clf2.predict(x_test_tranformed)
print("Accuracy LR",accuracy_score(y_test,y_pred1))
print("Accuracy DT",accuracy_score(y_test,y_pred2))
```

```python
x_transform=trf.fit_transform(x)
clf1=LogisticRegression()
clf2=DecisionTreeClassifier()
print("LR",np.mean(cross_val_score(clf1,x_transform,y,scoring='accuracy',cv=10)))
print("DT",np.mean(cross_val_score(clf2,x_transform,y,scoring='accuracy',cv=10)))
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
stats.probplot(x_train['Fare'],dist="norm",plot=plt)
plt.title("Fare before Log")
plt.subplot(122)
stats.probplot(x_train_tranformed['Fare'],dist="norm",plot=plt)
plt.title("Fare after Log")
plt.show()
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
stats.probplot(x_train['Age'],dist="norm",plot=plt)
plt.title("Age before Log")
plt.subplot(122)
stats.probplot(x_train_tranformed['Age'],dist="norm",plot=plt)
plt.title("Age after Log")
plt.show()
```

```python
trf2=ColumnTransformer([('log',FunctionTransformer(np.log1p),['Fare'])],remainder='passthrough')
```

```python
x_train_transformed2=trf2.fit_transform(x_train)
x_test_transformed2=trf2.transform(x_test)
```

```python
clf1=LogisticRegression()
clf2=DecisionTreeClassifier()
clf1.fit(x_train_transformed2,y_train)
clf2.fit(x_train_transformed2,y_train)
y_pred1=clf1.predict(x_test_transformed2)
y_pred2=clf2.predict(x_test_transformed2)
print("Accuracy LR",accuracy_score(y_test,y_pred1))
print("Accuracy DT",accuracy_score(y_test,y_pred2))
```

```python
x_transform2=trf2.fit_transform(x)
clf1=LogisticRegression()
clf2=DecisionTreeClassifier()
print("LR",np.mean(cross_val_score(clf1,x_transform2,y,scoring='accuracy',cv=10)))
print("DT",np.mean(cross_val_score(clf2,x_transform2,y,scoring='accuracy',cv=10)))
```

```python
def apply_transform(transform):
    x=df.iloc[:,1:3]
    y=df.iloc[:,0]
    trf=ColumnTransformer([('log',FunctionTransformer(transform),['Fare'])],remainder='passthrough')
    x_trans=trf.fit_transform(x)
    clf=LogisticRegression()
    print("Accuracy",np.mean(cross_val_score(clf,x_trans,y,scoring='accuracy',cv=10)))
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    stats.probplot(x['Fare'],dist="norm",plot=plt)
    plt.title("Fare before transform")
    plt.subplot(122)
    stats.probplot(x_trans[:,0],dist="norm",plot=plt)
    plt.title("Fare after transform")
    plt.show()
```

```python
apply_transform(lambda x:x)
```
