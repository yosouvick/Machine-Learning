# Distplot 

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
titanic['age'].fillna(titanic['age'].mean(),inplace=True)
```

```python
titanic.isnull().sum()
```

# Plotting a distplot

```python
sns.distplot(titanic['age'])
```

# plotting a histogram


```python
sns.distplot(titanic['age'],kde=False,bins=10)
```

# Plotting the only KDE

```python
sns.distplot(titanic['age'],hist=False)
```

# Plotting only the Rugplot

```python
sns.distplot(titanic['age'],kde=False,hist=False,rug=True)
```
