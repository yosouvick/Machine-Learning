```python
def run_scored(batsman):
    bat=df1[df1['batsman']==batsman]
    return bat.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
run_scored('V Kohli')
```

```python
#isin()
#find te most destructive death over of a batsman in ipl
#strike rate=(no of runs/no of balls)*100
#Min 200 balls played by batsman in over 16-20
```

```python
#batsman played between the over 16-20
mask=df1['over']>15
death_over=df1[mask]
```

```python
all_batsman=death_over.groupby('batsman')['batsman_runs'].count() #total no of ball played by batsman
x=all_batsman>200
batsman_list=all_batsman[x].index.tolist()
```

```python
batsman_list
```

```python
death_over['batsman'].isin(batsman_list)
```

```python
#runs scored by all 43 batsman
#balls played by all 43 batsman
Final=death_over[death_over['batsman'].isin(batsman_list)]
```

```python
runs=Final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls=Final.groupby('batsman')['batsman_runs'].count()
```

```python
sr=(runs/balls)*100
```

```python
sr.sort_values(ascending=False).head(5)
```
